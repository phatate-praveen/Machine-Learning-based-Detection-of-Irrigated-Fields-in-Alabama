```python
# Append Labels function
def append_labels(combined_data, al_labels, intreset_geo):
    '''
    Take the image, labels and roi and creates the labels as a Band and appends it to the image
    input: Image, Labels and region of interest geometry
    Output: Return and Image
    
    '''
    single_lbl = al_labels.filterBounds(intreset_geo)
    al_irri_coor = single_lbl.geometry().coordinates()
    
    def irri_labels(ls):
        geo = ee.Geometry.Polygon(ls)
        ft = ee.Feature(geo, {'class': 1})
        return ft
    
    al_irri_ft = ee.FeatureCollection(al_irri_coor.map(irri_labels))

    # Label as non irrigated
    al_coords_geo = ee.Geometry.MultiPolygon(al_irri_coor.map(lambda ls: ee.Geometry.Polygon(ls)))
#     print(type(al_coords_geo))
    al_non_irri_coord = intreset_geo.difference(al_coords_geo, ee.ErrorMargin(0.001))
    al_non_irri = ee.Feature(al_non_irri_coord, {'class': 0})
    al_non_irri_ft = ee.FeatureCollection(al_non_irri)

    al_label_feature = al_irri_ft.merge(al_non_irri_ft)
    al_labels_img = al_label_feature.reduceToImage(
                      properties = ['class'],
                      reducer = ee.Reducer.first()
                    ).rename("label");


    al_featureStack = ee.Image.cat([combined_data, al_labels_img]).float()
   
    return al_featureStack

def get_cdl(season_year, roi):
    '''
    Take the year and roi and creates the cultivated as a Band
    input: year and region of interest geometry
    Output: Return an Image with cultivated as band
    
    '''
    dataset = (ee.ImageCollection('USDA/NASS/CDL').filter(ee.Filter.date(season_year[0], season_year[1]))
                  .mosaic()
                  .clip(roi))
    bandNames = dataset.bandNames();
    print('Image Band names:', bandNames.getInfo());
    cultivated = dataset.select('cultivated').eq(2)
    return cultivated
def cloudMask(image):
    '''
    Take the landsat image and applies cloud MAsk
    input: Landsat Image
    Output: Return the image after applying cloud mask
    
    '''
    qa = image.select('QA_PIXEL')
    # If the cloud bit (5) is set and the cloud confidence (7) is high
    # or the cloud shadow bit is set (3), then it's a bad pixel.
    cloud = qa.bitwiseAnd(1<<3).And(qa.bitwiseAnd(1<<9)).Or(qa.bitwiseAnd(1<<4))
    # Remove edge pixels that don't occur in all bands
    mask2 = image.mask().reduce(ee.Reducer.min());
    opticalBands = image.select('SR_B.').multiply(0.0000275).add(-0.2).copyProperties(image, ['system:time_start']);
#         masked_image = image.updateMask(cloud.not()).updateMask(mask2);
    return image.addBands(opticalBands, None, True).updateMask(cloud.Not()).updateMask(mask2)


def get_modis_collection(projection, period, intreset_geo):
    '''
    Generates modis time series data
    input: Takes the projection of landsat for bilinear smapling and growing season and region of interest
    Output: Return the modis timeseries Image as NDVI and EVI for every period in the growing season
    
    '''
#     print(period)
    modis16day = ee.ImageCollection("MODIS/006/MOD13Q1").filterBounds(intreset_geo)
#     projection = modis16day.mosaic().projection()
#     print('modis16day', projection.getInfo())
    modislst_timeSeries = ee.ImageCollection(modis16day
                    .filterDate(period[0], period[1])
                    .select('NDVI', 'EVI')
                    .map(lambda image: image.multiply(0.0001))
                    .map(lambda image : image.reproject(crs = projection, scale= 30.0).resample('bilinear'))
#                     .map(lambda image : image.resample('bilinear'))
                    )
    print(modislst_timeSeries.size().getInfo())
#     print("modis_image", ee.Image(modislst_timeSeries.get(0)).projection().getInfo())
    def mergeBands(image, previous):
        return ee.Image(previous).addBands(image)
    
    mergedMODIS_LSTCollection = ee.Image(modislst_timeSeries.iterate(mergeBands, ee.Image()))
    return mergedMODIS_LSTCollection


def get_landsat8_collection(growing_season, roi, al_geometry):
    '''
    Takes growing season, region of interest and alabama geometry and returns landsat 8 image
    input: growing season and region of interest
    Output: Return the Landsat Image with intrested bandas the projection which is used later for combining with MODIS
    
    '''
    landsat_8 = ee.ImageCollection("LANDSAT/LC08/C02/T1_L2").filterBounds(al_geometry)
    projection = landsat_8.mosaic().projection()
    landsat_image = ee.Image(landsat_8 
                          .filterDate(growing_season[0], growing_season[1])
                          .map(cloudMask)
                          .select(['SR_B2','SR_B3','SR_B4','SR_B5','SR_B6','SR_B7'])
                          .map(lambda image: image.clip(roi))
                        .mosaic()
                          ).rename(['BLUE', 'GREEN', 'RED', 'NIR', 'SWIR1', 'SWIR2']).addBands(ee.Image.pixelLonLat())
    return landsat_image, projection

def get_landsat5_collection(growing_season, roi, al_geometry):
    '''
    Takes growing season, region of interest and alabama geometry and returns landsat 5 image
    input: growing season and region of interest
    Output: Return the Landsat Image with intrested bandas the projection which is used later for combining with MODIS
    
    '''
    landsat_5 = ee.ImageCollection("LANDSAT/LT05/C02/T1_L2").filterBounds(al_geometry)
    projection = landsat_5.mosaic().projection()
    landsat_image = ee.Image(landsat_5 
                      .filterDate(growing_season[0], growing_season[1])
                      .map(cloudMask)
                      .select(['SR_B1','SR_B2','SR_B3','SR_B4','SR_B5','SR_B7'])
                      .map(lambda image: image.clip(roi))
                    .mosaic()
                      ).rename(['BLUE', 'GREEN', 'RED', 'NIR', 'SWIR1', 'SWIR2'])
    return landsat_image, projection
```
